This notebook will query date time information all websites of interest.

```python
from mwviews.api import PageviewsClient
import pandas as pd
import os
import pathlib
p = PageviewsClient(user_agent = 'voje@gmail.com')
%matplotlib inline
```

```python
path_to_links = pathlib.Path('../data/test15/wikilinks/')
path_to_revisions = pathlib.Path('../data/test15/revisions/')

pages = [f.name.split('.json')[0] for f in path_to_links.iterdir()]
```

```python
pageview_data = pd.DataFrame.from_dict(p.article_views('en.wikipedia', pages,agent='user', start=ultimate_start_date)
                       , orient='index')
```
